```python
import numpy as np
import pandas as pd
```

```python
x_train = []
x_test = []
y_train = []
y_test = []

with open('data.txt') as f:
    i = 0
    for line in f:
        if i not in [0,1,2,3,11,12]:
            x_train.append([np.float32(line.split(',')[0]), np.float32(line.split(',')[1])])
            y_train.append(np.float32(line.split(',')[2].strip()))
        else:
            x_test.append([np.float32(line.split(',')[0]), np.float32(line.split(',')[1])])
            y_test.append(np.float32(line.split(',')[2].strip()))
        i += 1

print(x_train)
print(y_train)
print(x_test)
print(y_test)
```

```python
class Perceptron:
    """
    1 Hidden Layer MLP Class
    
    Input:  x - 2 dimensions of both train params
            y - 1 dimensional true values
            lr - learning rate (default: 0.1)
    """
    def __init__(self, x, y, lr=0.1):
        self.x = x
        self.y = y
        self.input_size = 2
        self.w = self.initialization(self.input_size)
        self.b = self.initialization(1)
        self.a = 0
        self.lr = lr
        self.pred_y = []
    
    def initialization(self, size):
        return np.random.rand(size)
    
    def activation_function(self, z):
        if z > 0:
            return 1
        else:
            return -1
    
    def forward_propagation(self, x, test=None):
        output_layer = 0
        for i in range(self.input_size):
            z = self.w[i] * x[i]
            output_layer += z
        if test != None:
            self.pred_y.append(self.activation_function(output_layer + self.b))
        else:
            self.a = self.activation_function(output_layer + self.b)
    
    def error_function(self, y_true, output_layer):
        return y_true - output_layer
    
    def params_ouput_update(self, e, x):
        for i in range(len(self.w)):
            self.w[i] = self.w[i] + self.lr * e * x[i]
        self.b = self.b + self.lr * e
            
    def train(self):
        error_count_flag = 1
        j = 0
        while error_count_flag != 0:
            j += 1
            error_count = []
            for i in range(len(self.x)):
                self.forward_propagation(self.x[i])
                e = self.error_function(self.y[i], self.a)
                self.params_ouput_update(e, self.x[i])
                error_count.append(e)
            print(f'Epoch: {j}')
            if -1 not in error_count and -2 not in error_count:
                error_count_flag = 0
                print(f'error for each x: {error_count}')

    def test(self, x_test, y_test):
        for x in x_test:
            self.forward_propagation(x, test='test')
        print(f'y_true: {y_test}')
        print(f'y_pred: {self.pred_y}')
```

```python
model = Perceptron(x_train, y_train, lr=0.1)
model.train()
model.test(x_test, y_test)
```
